## Linear Regression with Single Variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('csv/homeprices.csv')
df
```

```python
plt.title('Home Prices')
plt.xlabel('area (in sq ft)')
plt.ylabel('price (US $)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300], [5000], [2600]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv('csv/areas.csv')
d.head()
```

```python
prices = reg.predict(d)
```

```python
d['prices'] = prices
```

```python
d.to_csv('csv/prediction.csv', index = False)
```

```python
plt.title('Home Prices')
plt.xlabel('area', fontsize = 20)
plt.ylabel('price', fontsize = 20)
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'b')
plt.show()
```
